```python
import pandas as pd
import matplotlib.pyplot as plt
```

Shared memory BLOCK_SIZE from time

```python
dict_avar = {'1024': {4: [60.6478],
                      8: [55.9676],
                      16: [20.444],
                      32: [17.1974],},
             '2048': {4: [98.5679],
                      8: [48.571],
                      16: [30.58681],
                      32: [25.6732]},
             '4096': {4: [458.557],
                      8: [221.379],
                      16: [129.891],
                      32: [121.612]},
             '8192': {4: [4112.67],
                     8: [1438.97],
                     16: [610.082],
                     32: [524.783],
                    }, 
             '16384': {4: [40843.2],
                     8: [12128],
                     16: [4131.47],
                     32: [3893.7],
                    }
            }
   
```

```python
result_dict_time = {}
for key, vals in dict_avar.items():
    if key not in result_dict_time:
        result_dict_time[key] = {}
    for k, v in vals.items():
        if k not in result_dict_time[key]:
            result_dict_time[key][k] = 0
        result_dict_time[key][k] = sum(dict_avar[key][k]) / len(dict_avar[key][k])
```

```python
df_avat = pd.DataFrame().from_dict(result_dict_time).T
for row in df_avat.iterrows():
    print(row[1].index)
    plt.plot(row[1].index, row[1])
plt.legend(labels=df_avat.index)
plt.title("Optimization O3")
plt.xlabel("Num. of process")
plt.ylabel("Time in seconds")
plt.show()
```

```python
result_dict_flops = {}
for key, vals in dict_avar.items():
    if key not in result_dict_flops:
        result_dict_flops[key] = {}
    for k, v in vals.items():
        if k not in result_dict_flops[key]:
            result_dict_flops[key][k] = 0
        res = sum(dict_avar[key][k]) / len(dict_avar[key][k])
        result_dict_flops[key][k] = (2 * int(key) ** 3) / ((res / 1000) * 10 ** 9)
        
```

```python
df_flops = pd.DataFrame().from_dict(result_dict_flops).T
df_flops
```

```python
for row in df_flops.iterrows():
    print(row[1].index)
    plt.plot(row[1].index, row[1])
plt.legend(labels=df_flops.index)
plt.title("Graphs")
plt.xlabel("BLOCK_SIZE")
plt.ylabel("GFLOPS")
plt.show()
```
